```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from rdkit.Chem import PandasTools

import tensorflow as tf
import pandas as pd
import rdkit, rdkit.Chem, rdkit.Chem.rdDepictor, rdkit.Chem.Draw
import networkx as nx


emb = pd.read_csv('./data/embl.tsv', sep =  '\t')
```

```python
emb.head()
```

```python
emb.info()
```

```python
emb.shape
```

```python
# Bioactivities are in nM
emb['Bioactivities'].isna().sum()
```

```python
emb_bio = emb.dropna(subset = ['Bioactivities'])
```

```python
emb_bio.shape
```

```python
emb_bio.head()
```

```python
emb['Smiles'].isna().sum()
```

```python
emb_bios = emb_bio.dropna(subset = ['Smiles'])
```

```python
emb_bios.shape
```

```python
emb_bios['Bioactivities'].max(), emb_bios['Bioactivities'].min()
```

```python
emb_bios = emb_bios[emb_bios['Bioactivities'] <= 1000]
```

```python
emb_bios.shape
```

```python
duplicates = emb_bios[emb_bios.duplicated(subset=['ChEMBL ID'])]
```

```python
duplicates.isna().sum()
```

```python
# change Bioactivities to IC50
emb_bios = emb_bios.rename(columns={'Bioactivities': 'IC50'})
```

```python
# add new column 'unit' for nM
emb_bios['Unit']='nM'
```

```python
# add new column of with pIC50 values. pIC50 in M
#https://projects.volkamerlab.org/teachopencadd/talktorials/T001_query_chembl.html#
import math
def convert_ic50_to_pic50(IC50_value):
    pIC50_value = 9 - math.log10(IC50_value)
    return pIC50_value
```

```python
emb_bios["pIC50"] = emb_bios.apply(lambda x: round(convert_ic50_to_pic50(x.IC50), 2), axis=1)
```

```python
emb_bios.head()
```

```python
fig, (axes) = plt.subplots(2,3, figsize = (12,8))

axes[0,0].hist(emb_bios['Molecular Weight'], color = 'navy', bins = 100)
axes[0,0].set_xlabel('Molecular Weight, Da', fontsize = 13)
axes[0,0].set_ylabel('Frequency', fontsize = 13)
axes[0,1].hist(emb_bios['AlogP'], color = 'orange', bins = 100)
axes[0,1].set_xlabel('LogP', fontsize = 13)
axes[0,2].hist(emb_bios['HBA'], color = 'skyblue', bins = 15)
axes[0,2].set_xlabel('No. hydrogen bond acceptors', fontsize = 13)
axes[1,0].hist(emb_bios['HBD'], color = 'rosybrown', bins = 15)
axes[1,0].set_ylabel('Frequency', fontsize = 13)
axes[1,0].set_xlabel('No. of hydrogen bond donors', fontsize = 13)
axes[1,1].hist(emb_bios['#RO5 Violations (Lipinski)'], color = 'gold', bins = 20)
axes[1,1].set_xlabel('No. of RO5 violations', fontsize = 13)
axes[1,2].hist(emb_bios['pIC50'], color = 'teal', bins = 100)
axes[1,2].set_xlabel('pIC50, M', fontsize = 13)
fig.suptitle('Physicochemical properties of approved drugs', fontweight="bold", fontsize=20);
```

```python
# Seelect the columns that are required for further analysis
emb_bios = emb_bios.reindex(columns=['ChEMBL ID', 'IC50', 'Unit', 'Smiles', 'pIC50',])
```

```python
emb_bios.head()
```

```python
emb_bios.to_csv('./data/approved.csv', index = False)
```
